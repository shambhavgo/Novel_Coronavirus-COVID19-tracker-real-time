```python
#String for Current Time and Date
from datetime import datetime
now = datetime.now()
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
```

```python
#Initializing Important Functions
def confirmed(place_name):
    for state in response['state_wise']:
        if place_name.lower()==state.lower():
            return int(response['state_wise'][state]['confirmed'])
        else:
            for city in response['state_wise'][state]['district']:
                if city.lower() == place_name.lower():
                    return int(response['state_wise'][state]['district'][city]['confirmed'])
                
                
def active(place_name):
    
    for state in response['state_wise']:
        if place_name.lower()==state.lower():
            return int(response['state_wise'][state]['active'])
        else:
            for city in response['state_wise'][state]['district']:
                if city.lower() == place_name.lower():
                    return int(response['state_wise'][state]['district'][city]['active'])

def recovered(place_name):
    
    for state in response['state_wise']:
        if place_name.lower()==state.lower():
            return int(response['state_wise'][state]['recovered'])
        else:
            for city in response['state_wise'][state]['district']:
                if city.lower() == place_name.lower():
                    return int(response['state_wise'][state]['district'][city]['recovered'])

def deaths(place_name):
    
    for state in response['state_wise']:
        if place_name.lower()==state.lower():
            return int(response['state_wise'][state]['deaths'])
            
        else:
            for city in response['state_wise'][state]['district']:
                if city.lower() == place_name.lower():
                    return int(response['state_wise'][state]['district'][city]['deceased'])
                
def last_updated(place_name):
    
    for state in response['state_wise']:
        if place_name.lower()==state.lower():
            return response['state_wise'][state]['lastupdatedtime']
        else:
            for city in response['state_wise'][state]['district']:
                if city.lower() == place_name.lower():
                    return response['state_wise'][state]['lastupdatedtime']
```
